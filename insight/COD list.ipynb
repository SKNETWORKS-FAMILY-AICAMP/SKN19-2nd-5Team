## 분류 이유
### 1. 암 직접 사망 vs 관리 가능한 사망 원인 구분

COD 데이터에는 암 자체 사망 원인과 비(非)암성 원인이 혼재되어 있음.

그중 패혈증, 폐렴, 감염, 당뇨 합병증 등은 병원 집중 케어를 통해 예방·완화가 가능한 합병증성 사망 원인임.

따라서 Complications 그룹으로 별도 분류함.

### 2. 병원 서비스/비즈니스 관점

Cancer 그룹: 치료 성과와 직접 연결 → 의료기술·치료제 성과 지표.

Complications 그룹: 관리·간호 품질과 연관 → 병원 질 관리(QoC) 타깃.

Other 그룹: 질환이지만 병원 개입 여지가 적은 원인 → 직접적인 품질 관리 영역 아님.

External 그룹: 사고, 타살, 불명 사망 → 질병과 무관, 통계상 분리 필요.

Suicide 그룹: 자살/자가손상 → 별도 관리 및 정신건강 개입 필요.

### 3. Complications 주요 예시

패혈증, 폐렴, 감염: 면역저하 암환자에서 흔히 발생.

간경화, 신부전: 항암치료·방사선 부작용으로 악화.

COPD, 당뇨, 고혈압, 알츠하이머: 고령 암환자 동반질환으로 치료 중 사망 기여.

### 분류 기준

Cancer = 암 자체 원인 사망임.

Complications = 병원 집중 케어로 예방·완화 가능한 합병증성 사망임.

Other = 직접 개입하기 어려운 질환성 사망임.

External = 질병과 무관한 외인사·불명 사망임.

Suicide = 자살/자가손상 사망임.

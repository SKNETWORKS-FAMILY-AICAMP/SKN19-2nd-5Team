```python
# 프로젝트 루트를 sys.path에 추가하여 'modules' 임포트 가능하게 설정
from pathlib import Path
import sys
_CWD = Path.cwd()
_ROOT = _CWD.parent if (_CWD.name == 'insight' and (_CWD.parent / 'modules').exists()) else _CWD
if str(_ROOT) not in sys.path:
    sys.path.insert(0, str(_ROOT))
print('Project root:', _ROOT)
```

```python
# 필요한 모듈 임포트 및 시각화 실행
from modules.DataAnalysis import show_graph
import pandas as pd
from pathlib import Path

# encoded_dataset_COD.csv를 로드하여 DataFrame으로 전달
candidates = [
    Path('data/encoded_dataset_COD.csv'),
    Path('../data/encoded_dataset_COD.csv'),
    Path('data/EDA/encoded_dataset_COD.csv'),
    Path('../data/EDA/encoded_dataset_COD.csv'),
]
df = None
for p in candidates:
    if p.exists():
        df = pd.read_csv(p)
        print(f'Loaded: {p}')
        break
if df is None:
    raise FileNotFoundError('encoded_dataset_COD.csv not found in expected paths')

show_graph(df)
```
